```python
import pandas as pd
```

```python
data = pd.read_csv("House Price India.csv") 
```

```python
data.head()
```

```python
data.info()
```

```python
stats = data['Price'].describe().reset_index()
stats['Price'] = round(stats['Price'],2) 
```

```python
stats
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

```python
# for na values
data.dropna(inplace= True)
# for duplicated values
data.drop_duplicates(inplace= True)
```

```python
data.columns
```

```python
data.head()
```

```python
import matplotlib.pylab as plt
```

```python
data.groupby("condition of the house")["Price"].mean().sort_values(ascending=False).plot(kind= "bar", figsize=(8,5))
plt.title("average price of house condition")
plt.xlabel("condition of the house")
plt.ylabel("average price")
plt.show()
```

```python
data.columns
```

```python
X = data[['number of bedrooms', 'number of bathrooms',
       'living area','condition of the house', 'Number of schools nearby']]
y = data['Price']
```

```python
X
```

```python
y
```

```python
data.shape
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2 )
```

```python
X_test.shape
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import DecisionTreeRegressor
```

```python
param_grid = {
  'criterion' : ["squared_error", "friedman_mse", "absolute_error"],
  "splitter" : ["best", "random"],
  "max_depth" : [None, 10, 20, 30, 40, 50],
  "min_samples_split" : [2,5,10],
  "min_samples_leaf" : [1,2,4]
}
```

```python
tree_model = DecisionTreeRegressor()
```

```python
grid_tree = GridSearchCV(estimator= tree_model, param_grid= param_grid)
```

```python
grid_tree.fit(X_train, y_train)
```

```python
grid_tree.best_params_
```

```python
grid_tree.best_score_ 
```

```python
tree_pred = grid_tree.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error
print(mean_squared_error(y_test, tree_pred))
print(mean_absolute_error(y_test, tree_pred))
```

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)
```

```python
pred_lin = lin_model.predict(X_test)
```

```python
print(mean_squared_error(y_test, pred_lin))
print(mean_absolute_error(y_test, pred_lin))
```

```python
from sklearn.ensemble import RandomForestRegressor
rf_model = RandomForestRegressor()
```

```python
param_gridrfr = {
  "n_estimators" : [2,3,4,5,6,7,8,9,10],
  "max_depth" : [5,10,15]
}
gridrfr = GridSearchCV(rf_model,param_gridrfr)
```

```python
gridrfr.fit(X_train, y_train)
```

```python
gridrfr.best_params_
```

```python
rfrprediction = gridrfr.predict(X_test)
```

```python
print(mean_absolute_error(y_test, rfrprediction))
print(mean_squared_error(y_test, rfrprediction))
```
